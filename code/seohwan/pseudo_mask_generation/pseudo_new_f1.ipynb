```python
import os
import shutil

# # 상위 20%를 추출하기 위한 기준값 계산
# top_20_percent_threshold = df['scored_geo'].quantile(0.8)

source_dir = '/home/team1/ddrive/team1/SAM_WSSS_CAM/SAM_WSSS_CAM_0705/code/pseudo_mask_generation/outputs_vit_b_internel_iter2_soft/train_pseudo_masks_iter2_point_prompt_proba'
dest_dir = '/home/team1/ddrive/team1/SAM_WSSS_CAM/SAM_WSSS_CAM_0705/code/pseudo_mask_generation/outputs_vit_b_internel_iter2_soft/train_pseudo_masks_point_prompt_os_filter_iter2_trash'

# 목적지 디렉토리가 없으면 생성합니다.
os.makedirs(dest_dir, exist_ok=True)
num = 0
# 데이터프레임의 파일명을 순회합니다.
for index, row in df_sorted_4_filter.iterrows():
    file_name = row['file']
    source_file_path = os.path.join(source_dir, file_name)
    dest_file_path = os.path.join(dest_dir, file_name)
    
    # 파일이 존재하는지 확인하고 복사합니다.
    if os.path.exists(source_file_path):
        shutil.copy2(source_file_path, dest_file_path)
        num += 1
        print(f"{file_name}이(가) 성공적으로 복사되었습니다.")
    else:
        print(f"{file_name}을(를) 찾을 수 없습니다.")

print(f'-----------------상위 30% 파일 복사 완료------------------')
print(num)
```
